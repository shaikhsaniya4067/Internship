```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\rutuj\Downloads\DSDBA\Heart_dataset.csv", index_col=0)

df['Age'].plot(kind='hist')
plt.title("Histogram of Age")
plt.show()


print("Shows distribution of age. Most patients are in middle-age group (clustered values).")
```

```python
df.plot(kind='box')
plt.title("Box Plot")
plt.show()
print("Used to detect outliers. Some attributes like cholesterol may have extreme values.")
```

```python
df['Age'].plot()
plt.title("Line Plot of Age")
plt.show()
print("Shows trend of age values across dataset. No specific trend (random distribution).")
```

```python
df.mean(numeric_only=True).plot(kind='bar')
plt.title("Bar Plot of Mean Values")
plt.show()
print("Compares average values of all features. Helps identify which attribute has higher magnitude.")
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart of Sex")
plt.ylabel('')
plt.show()
print("Shows proportion of male vs female patients.")
```

```python
plt.scatter(df['Age'], df['Chol'])
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Scatter Plot")
plt.show()
print("Shows relationship between age and cholesterol. Weak or no clear correlation.")
```

```python
df.head(10).plot.area()
plt.title("Area Plot")
plt.show()
print("Shows cumulative contribution of features over samples.")
```

```python
df.plot(kind='density')
plt.title("Density Plot")
plt.show()
print("Shows smooth distribution of data. Helps understand data concentration.")
```
